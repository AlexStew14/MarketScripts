```python
import yfinance as yf
import pandas as pd
import plotly.graph_objects as go
from datetime import datetime
```

```python
stock = yf.Ticker("AEO")

all_chains = [stock.option_chain(x) for x in stock.options]
```

```python
all_calls = [x.calls for x in all_chains[:2]]
days_until_expiration = [(datetime.strptime(x, '%Y-%m-%d') - datetime.today()).days for x in stock.options[:2]]

all_calls_with_expirations = []
for c, d in zip(all_calls, days_until_expiration):
    c['daysUntilExpiration'] = d
    all_calls_with_expirations.append(c)
```

```python
all_calls_df = pd.concat(all_calls_with_expirations)
```

```python
print(all_calls_df.info())
```

```python
import numpy as np
print(all_calls_df.impliedVolatility.shape)
z = np.tile(all_calls_df.impliedVolatility.values, (all_calls_df.impliedVolatility.shape[0], 1))
print(z)
```

```python
x = all_calls_df.daysUntilExpiration.values
y = all_calls_df.strike.values
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])
fig.update_layout(title='Implied Volatility', autosize=False,
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))
fig.show()
```
