```python
from td.client import TDClient
import pandas as pd
```

```python
from api_info import client_id, redirect_uri

TDSession = TDClient(
    client_id=client_id,
    redirect_uri=redirect_uri,
    credentials_path='./td_state.json'
)

# Login to the session
TDSession.login()
```

```python
accnt = TDSession.get_accounts(fields=['positions'])[0]
```

```python
positions = accnt['securitiesAccount']['positions']
position_tickers = []

invalid_chars = set("_0123456789")

for p in positions:
    ticker = p['instrument']['symbol']
    if (not invalid_chars.intersection(set(ticker))):
        position_tickers.append(ticker)

print(position_tickers)
```

```python
all_positions = []
for t in position_tickers:
    all_positions.append(TDSession.get_price_history(t, extended_hours=False, period=2, period_type='year', frequency=1, frequency_type='daily'))
```
